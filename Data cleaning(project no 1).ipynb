```python
#importing py laibraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#import dataset or loading dataset
df = pd.read_csv('movies.csv')
df.set_index('name', inplace=True)
```

```python
#printing dataset
df.head()
```

```python
#Multiplying year and score
#this is a demostration on how to Multiplying columns
df['year times score'] = df['year'] * df['score']
df.head()
```

```python
#list all the columns on the dataset
movies = list(df.columns)
movies
```

```python
#check for missing(NaN)
df.isnull().mean()
df
```

```python
#checking for missing value in per form from columns
#which shows that budget and gross has most missing values
print(df.isnull().mean())
print('')
```

```python
#sorting out the missing values for the numerical dataset for budget and gross to = 0
loops = ['budget', 'gross',]
for i in loops:
    df[i] = df[i].fillna('0')
```
